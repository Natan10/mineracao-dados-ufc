# <span style="color:blue"> 3. Pré-Processando os datasets </span>

Limpando a coluna `Age` para que ela possua apenas números, e a coluna `Rotten Tomatoes`, para que fica similar a `Age`

```python
## Limpeza na coluna de idade, removendo o '+'

for i in range(len(df_movies['Age'])):
    aux = df_movies.loc[i,'Age']
    if aux != np.nan or aux != 'all' or aux != '':
        aux = str(aux)
        if '+' in aux:
            df_movies.loc[i,'Age'] = aux.strip().split('+')[0]
                        

for i in range(len(df_series['Age'])):
    aux = df_series.loc[i,'Age']
    if aux != np.nan or aux != 'all' or aux != '':
        aux = str(aux)
        if '+' in aux:
            df_series.loc[i,'Age'] = aux.strip().split('+')[0]
                                        
```

```python
## Limpeza na coluna do Rotten Tomatoes no Filmes

for i in range(len(df_movies['Rotten Tomatoes'])):
        aux = df_movies.loc[i,'Rotten Tomatoes']
        if aux != np.nan or aux != '':
            aux = str(aux)
            if '%' in aux:
                df_movies.loc[i,'Rotten Tomatoes'] = aux.strip().split('%')[0]  
                
for i in range(len(df_series['Rotten Tomatoes'])):
        aux = df_series.loc[i,'Rotten Tomatoes']
        if aux != np.nan or aux != '':
            aux = str(aux)
            if '%' in aux:
                df_series.loc[i,'Rotten Tomatoes'] = aux.strip().split('%')[0]          

## Deixa os valores como floats de 0 a 10, assim como o IMDb                
df_movies['Rotten Tomatoes'] = df_movies['Rotten Tomatoes'].apply(lambda x: float(x)/10)                  
df_series['Rotten Tomatoes'] = df_series['Rotten Tomatoes'].apply(lambda x: float(x)/10)                  
```

### A coluna 'type' no dataset de series está com nomeação diferente , então tem-se de modificar seu nome para 'Type', para ficar igual nos dois datasets.
### E renomear a coluna 'Unnamed: 0' do df_series para 'ID', para ambos os dataframes ficarem com a uma coluna nomeada ID.

```python
df_series.rename(columns={'Unnamed: 0': 'ID', 'type': 'Type'},inplace=True)
```

### Por fim, ajeitar o ID das séries para ficar começar a partir do último ID dos filmes, e assim não haver IDs repetidos

```python
numMovies = len(df_movies.index)
df_series['ID']=df_series.apply(lambda x: x['ID']+numMovies+1,axis=1)
```

```python
# número de filmes em portugues no Prime Videos
pv_pt = df_languages[pt].iloc[df_prime_movies['ID']-1]
num_pv_pt = pv_pt[(pv_pt[pt[0]] == 1) | (pv_pt[pt[1]] == 1)].count()[0]
# número de filmes em portugues no Hulu
hulu_pt = df_languages[pt].iloc[df_hulu_movies['ID']-1]
num_hulu_pt = hulu_pt[(hulu_pt[pt[0]] == 1) | (hulu_pt[pt[1]] == 1)].count()[0]
# número de filmes em portugues na Netflix
net_pt = df_languages[pt].iloc[df_netflix_movies['ID']-1]
num_net_pt = net_pt[(net_pt[pt[0]] == 1) | (net_pt[pt[1]] == 1)].count()[0] 
# número de filmes em portugues na Disney+
disney_pt = df_languages[pt].iloc[df_disney_movies['ID']-1]
num_disney_pt = disney_pt[(disney_pt[pt[0]] == 1) | (disney_pt[pt[1]] == 1)].count()[0]

fig,axes = plt.subplots(nrows = 1,ncols = 1,figsize=(8,5),constrained_layout=True)
fig.suptitle('Quantidade de Filmes em Português de cada Plataforma')

x = ['Prime Video', 'Hulu','Netflix','Disney+']
y = [num_pv_pt, num_hulu_pt, num_net_pt, num_disney_pt]
axes.bar(x, y, color = ['blue', 'green', 'red', 'purple'])
axes.set_title('Prime Video')
plt.show()
```

```python
country_imdb_mean = []

for i in listCountries:
    listaAlvo = list(df_countries[df_countries[i]==1]['ID'].iloc[:]-1)
    aux = df_movies.iloc[listaAlvo,:]
    country_imdb_mean.append(aux['IMDb'].mean())

num_Filmed = []
for i in listCountries:
    filmed = df_countries[i].sum()
    num_Filmed.append(filmed)    

# fazendo uma série e arredondando para 1 casa decimal
country_imdb_mean = pd.Series(country_imdb_mean)
country_imdb_mean = country_imdb_mean.apply(lambda x: round(x,1))

aux = np.array([listCountries,country_imdb_mean,num_Filmed]).T
data_countries = pd.DataFrame(aux,columns='country Imdb num_Filmed'.split())

page_code = requests.get("https://countrycode.org/").content
soup = BeautifulSoup(page_code, 'html.parser')
tags = soup.find_all('tbody')
tr_tag = tags[0].find_all('tr')
name_code = []

for i in tr_tag:
    aux = i.find_all('td') 
    name_code.append([aux[0].get_text(),aux[2].get_text().split('/')[1]])
    
code = ['']*len(data_countries['country'])


for i in range(len(name_code)):
    for j in range(len(listCountries)):
        if name_code[i][0] == listCountries[j]:
            code[j] = name_code[i][1]
        pass
    
data_countries['code'] =  pd.Series(code)
data_countries = data_countries[data_countries['Imdb'] != 'nan']
data_countries = data_countries[data_countries['code'] != '']
data_countries = data_countries.sort_values('Imdb',ascending = False)

imdb = list(data_countries.iloc[:,1])
imdb = pd.Series(imdb)
imdb = imdb.apply(lambda x: float(x))

filmed = list(data_countries.iloc[:,2])
filmed = pd.Series(filmed)
filmed = filmed.apply(lambda x: float(x))
filmed = filmed.apply(lambda x: np.log10(x))
```
